## First order lineshape

Starting right off in Liouville space is hard and annoying. First, I will try to calculate a first-order response for a three level system, levels a,b, and c. Their energies are given by $\epsilon_i$.

To do this, start with the Lioville space expression for susceptibility, $\chi^{(1)} = - \frac{1}{\hbar} \left\langle\langle V\right | \mathscr G(\omega) \mathscr V 
\left| \rho(-\infty) \rangle\right\rangle$. 

Initial $\rho$ is given by $\left| \rho \rangle\right\rangle = \sum_a P(a) \left| aa \rangle\right\rangle$ where $P(a)$ is the population given by Boltzmann distribution. I'll set this to $1$ for $\left| aa \rangle\right\rangle$ eventually and zero for all other states. Expanding $\mathscr L$ using the Hamiltonian energies, we get its eigenvalues: $\mathscr L \left | aa \rangle\right\rangle = \hbar \omega_{ab}\left| aa \rangle\right\rangle$. This comes from the definition, $\mathscr L \left| ab \rangle\right\rangle \equiv \mathscr L \left | a \right\rangle \left\langle b \right | = H \left | a \right\rangle \left\langle b \right | - \left | a \right\rangle \left\langle b \right | H = (E_n - E_m) \left | a \right\rangle \left\langle b \right | = \hbar\omega_{nm}\left | a \right\rangle \left\langle b \right | \equiv \hbar\omega_{nm}\left| ab \rangle\right\rangle $.

Next, we also need the frequency space Liuoville Green function. Because we know the eigenvalues in Lioville space and by definition $\mathscr G (\omega) \equiv \sum_{nm} \frac{\left| nm \rangle\right\rangle \left\langle\langle nm \right|}{\omega - \frac{\mathscr L}{\hbar} + i\epsilon}$, we can just replace the function of an operator with a function of its eigenvalue. (I think) because there are no permanent dipoles in $V$, the Green function is diagonal and given by 

$$\mathscr G (\omega) = \sum_{cd} \frac{\left| cd \rangle\right\rangle \left\langle\langle cd \right|}{\omega - \omega_{cd} + i\epsilon}$$

Substituting this into the original expression yields 

$$\chi^{(1)} = - \frac{1}{\hbar} \sum_{acd} P(a) \frac{\langle\langle V \left| cd \rangle\right\rangle \left\langle\langle cd \right| \mathscr V \left| aa \rangle\right\rangle}{\omega - \omega_{cd} + i\epsilon}$$ 

Evaluating the dipole moments in the same way as $\mathscr L$ above yields $\left\langle\langle cd \right|\mathscr V \left| aa \rangle\right\rangle \equiv \text{Tr}\left[\left|d\right\rangle \left\langle a \right | [V, \left|a\right\rangle \left\langle a\right|]\right] = V_{ca}\delta_{da} - V_{da}^* \delta_{ca}$. This also used the hermiticity of $V$. (Somehow) this means that $\mathscr V \left| aa \rangle\right\rangle = \sum_b \left [ \mu_{ba} \left| ab \rangle\right\rangle - \mu_{ab} \left| ab \rangle\right\rangle\right]$ and we can replace summation over c,d by summation over b. Substituting yields

$$\chi^{(1)}(\omega) = - \frac{1}{\hbar} \sum_{ab} P(a) \left \{\frac{\langle\langle V \left| ba\rangle\right\rangle \left\langle\langle ba \right| \mathscr V \left| aa \rangle\right\rangle}{\omega - \omega_{ba} + i\epsilon} + \frac{\langle\langle V \left| ab\rangle\right\rangle \left\langle\langle ab \right| \mathscr V \left| aa \rangle\right\rangle}{\omega - \omega_{ab} + i\epsilon} \right \}$$ 

The Liouville space elements are all basically a permutation of $\mu_{ab}$ or $\mu_{ba}^*$. This gives the final expression 

$$\chi^{(1)}(\omega) = - \frac{1}{\hbar} \sum_{ab} P(a) \left|\mu_{ab}\right|^2 \left \{\frac{1}{\omega - \omega_{ba} + i\epsilon} - \frac{1}{\omega - \omega_{ab} + i\epsilon} \right \}$$

The signal in an ideal frequency-resolved experiment in the limit of short pulses is proportional to $\chi^2$. This basically looks like this:

````python
from itertools import product
from __future__ import division

# Generate all elements of rho at low temperature
# Only aa is filled

levels = 'abc'

w = Symbol(r'\omega', real=True)
e = Symbol(r'\epsilon', constant=True)
hbar = Symbol(r'\hbar', constant=True, real=True)

x,y = symbols('x,y', dummy=True)

G = lambda x,y: 1/(w - Symbol(r'\omega_{:s}'.format(x + y), ) + I*e)

rho = []
chi = S(0)
for a,b in product(levels, levels):
    #rho.append(Symbol(r'\rho_{}'.format(a+b)))
    chi += - Symbol(r'\rho_{}'.format(a+a))*Abs(Symbol(r'\mu_{}'.format(a+b), real=True))**2*( G(a,b) - G(b,a) )
    #print chi
chi *= 1/hbar
display(chi)
````

````python
# Substitute values into the expressions
# Low temperature limit, so all but rho_aa are 0
subs = {Symbol(r'\rho_aa'): S(1)}
subs.update({Symbol(r'\rho_{}'.format(a+a)): S(0) for a in 'bc' })

chi = chi.subs(subs, simultaneous=True)
display(chi)

# Put in values of dipole moments and energies
subs = {Symbol(r'\epsilon', constant=True): S(0), 
    Symbol(r'\mu_ab', real=True): S(1), 
    Symbol(r'\mu_ac', real=True): Rational(1,2), 
    Symbol(r'\omega_ba'): S(2),
    Symbol(r'\omega_ab'): -S(2),
    Symbol(r'\omega_ac'): S(14),
    Symbol(r'\omega_ca'): -S(14),
    hbar: S(1),
    w: Symbol('x'),
}
chi = chi.subs(subs)
Is = factor(expand(chi*chi));
````

## First order with relaxation

Basically, we have to add bath terms to get relaxation. Take simple reduced equations of motion,

$$
\begin{eqnarray}
\dot\rho_{vv'} & = & (-i \omega_{vv'} - \Gamma_{vv'})\rho_{vv'} \\
\Gamma_{vv'} &=& \frac{1}{2}(\gamma_v + \gamma_{v'} ) + \hat\Gamma_{vv'}
\end{eqnarray}
$$

The new frequency domain Green function is given by

$$\mathscr G(\omega) = \sum_{vv'}  \frac{\left| v v' \rangle\right\rangle \left\langle\langle v v' \right|}{\omega - \omega_{vv'} + i\Gamma_{vv'}}$$

After the same song and dance the susceptibility is given by 

$$\chi^{(1)} (\omega) = \frac{1}{\hbar} \sum_{ab} P(a) \frac{2 \left | \mu_{ab} \right |^2 \omega_{ba}}{\omega_{ba}^2 - (\omega + i\Gamma_{ab})^2}$$

The plot shows the real and imaginary parts of $\chi$.

````python
from itertools import product
from __future__ import division

# Generate all elements of rho at low temperature
# Only aa is filled

levels = 'abc'

w = Symbol(r'\omega', real=True)
e = Symbol(r'\epsilon', constant=True)
hbar = Symbol(r'\hbar', constant=True, real=True)

G = lambda x,y: Symbol(r'\omega_{:s}'.format(y+x))/(Symbol(r'\omega_{:s}'.format(y+x))**2 - (w + I*Symbol(r'\Gamma_{}'.format(x+y)))**2)

rho = []
chi = S(0)
for a,b in product(levels, levels):
    #rho.append(Symbol(r'\rho_{}'.format(a+b)))
    chi += Symbol(r'\rho_{}'.format(a+a))*Abs(Symbol(r'\mu_{}'.format(a+b), real=True))**2*G(a,b)
    #print chi
chi *= 1/hbar
#display(chi)

subs = {Symbol(r'\rho_aa'): S(1), Symbol(r'\omega_aa'): S(0)}
subs.update({Symbol(r'\rho_{}'.format(a+a)): S(0) for a in 'bc' })
chi = chi.subs(subs)
display(chi)

# Put in values of dipole moments and energies
subs = { \
    Symbol(r'\mu_ab', real=True): S(1), 
    Symbol(r'\mu_ac', real=True): Rational(1,2), 
    Symbol(r'\omega_ba'): S(2),
    Symbol(r'\omega_ca'): S(14),
    hbar: S(1),
    Symbol(r'\Gamma_ac'): S(1),
    Symbol(r'\Gamma_ab'): S(2),
    w: Symbol('x', real=True),
}
chi = chi.subs(subs)
chi_r = re(chi)
chi_im = im(chi)

chi_rn = lambdify(Symbol('x'), chi_r)
chi_imn = lambdify(Symbol('x'), chi_im)
# plot it
wn = linspace(0, 20, 1024)
figure(figsize=(12,8))
plot(wn, chi_rn(wn));
plot(wn, chi_imn(wn));
````

## Two-photon process

To quote from the paper, 
    
> Coherent 2n-photon processes and incoherent n-photon processes are both
> probing a similar 2n -- time-correlation function of the dipole operator. 
> The reason is that in a coherent process we calculate an ensemble averaged 
> amplitude whereas in an incoherent process we average the cross section (amplitude
> square) over the ensemble.

I'm not sure why Mukamel doesn't give a frequency-resolved expression of the process. I'll be detecting the signal in the time domain, so this isn't a problem for me, but it would be nice to see. I should derive this. Anyway, he takes the Liouville space expression for $\chi^{(3)}$ and then proceeds to manipulate it.

$$\chi^{(3)}(-\omega_4; \omega_1, -\omega_2, \omega_3) = \sum_{P(\omega_1, -\omega_2, \omega_3)} \sum_a P(a) \left\langle\langle V \right | \mathscr G(\omega_1 - \omega_2 + \omega_3) \mathscr V \mathscr G(\omega_1 - \omega_2) \mathscr V \mathscr G(\omega_1) \mathscr V \left | aa \rangle \right \rangle$$

He then invokes the factorization approximation, which basically means that he averages the Green function over the ensemble and replaces each $\mathscr G$ by its average. This gives a simplified expression for G. By the way, G is in Liouville space so it has implicit indexes. He calls the generalized lineshape function $I_{\mu\nu}$. In the short correlation time limit for the bath $g_{\mu\nu}(\tau) = \hat\Gamma_{\mu\nu}\left|\tau\right|$ 

$$I_{\mu\nu}(\omega) = - i \int_0^\infty d \tau \exp[i\omega \tau]\exp[-i\omega_{\mu\nu}\tau - \frac{1}{2}(\gamma_\mu + \gamma_\nu)\tau - g_{\mu\nu}(\tau)]$$

Mukamel then evaluates this expression for a four-level system in terms of the lineshape function and obtains the 4WM result of Bloembergen. For the two-photon experiment, this expression is reused, but he only explores a three-level absorber with levels abc, (where b is the virtual transition level), thus setting $b=d$ in $\chi^{(3)}$. In the two-photon process, the signal fields are $\omega_L$ and $\omega_s$, and the system absorbs energy from both. Further, $\omega_L$ should near resonance with the virtual transition $\omega_{ab}$. He then defines the detuning from the energy levels $\Delta_L$ and $\Delta_s$. Energy of $a \lt b \lt c$.

$\begin{eqnarray}
\Delta_L &=& \omega_L - \omega_{ba} \\
\Delta_s &=& \omega_{cb} - \omega_{s} \\
\end{eqnarray}$

The observable is the rate of change of population $c$, $\hat{I}(t, \Delta_L, \Delta_s) = -i \langle\langle cc | \mathscr V | \rho(t) \rangle\rangle$. The final expression in the impact limit and with the factorization approximation is

$$\begin{eqnarray}
\hat{I}(t; \Delta_L, \Delta_s) = (-i)^4 \int_{-\infty}^{t} d\tau_1\int_{-\infty}^{\tau_1} d\tau_2\int_{-\infty}^{\tau_2} d\tau_3 &I_{cb}&(t-\tau_1)I_{bb}(\tau_1 - \tau_2)I_{ab}(\tau_2 - \tau_3) \overbrace {\langle E_L(\tau_2) E_L(\tau_3)\rangle \langle E_s(\tau_2) E_s(\tau_3)\rangle}^{\text{field fluctuations}} \\
+\vphantom{\int}&I_{bc}&(t-\tau_1)I_{bb}(\tau_1 - \tau_2)I_{ab}(\tau_2 - \tau_3) \langle E_L(\tau_2) E_L(\tau_3)\rangle \langle E_s(\tau_2) E_s(\tau_3)\rangle\\
+\vphantom{\int}&I_{bc}&(t-\tau_1)I_{ac}(\tau_1 - \tau_2)I_{ab}(\tau_2 - \tau_3) \langle E_L(\tau_2) E_L(\tau_3)\rangle \langle E_s(\tau_2) E_s(\tau_3)\rangle
\end{eqnarray}
$$

The time-domain lineshape is given by 

$$I_{\nu\mu}(\tau) = \exp \left[-i (E_{\nu} - E_{\mu})\tau -\frac{1}{2}(\gamma_\nu + \gamma_\mu)\tau - g_{\nu\mu}(\tau)\right]$$

with the definitions

$\begin{eqnarray}
E_a &=& \epsilon_a + \omega_L + \omega_s \\
E_b &=& \epsilon_b + \omega_s \\
E_c &=& \epsilon_c
\end{eqnarray}$

Now, let's try to simulate it. I will substitute $g_{\mu\nu}(\tau) = \hat\Gamma_{\mu\nu}\left|\tau\right|$, and take $E_L$ to be perfect pulse so the fluctuations are zero, and correlation is perfect. Also, start integration from zero instead of $-\infty$. I'm also looking at only the imaginary part of $\chi^{(3)}$.

````python
# Define the lineshape function
tau =  Symbol(r'\tau', dummy=True)
tau1, tau2, tau3, t = symbols(r'\tau_1,\tau_2,\tau_3,t', dummy=True, real=True, positive=True)

Ivu = lambda v,u, tau: exp(-I*(Symbol('E_'+v) - Symbol('E_'+u))*tau - Rational(1,2)*(Symbol('\gamma_'+v) + Symbol('\gamma_'+u))*tau \
    - Symbol('\Gamma_'+v+u)*Abs(tau))

integrand = Ivu('c', 'b', t - tau1)*Ivu('b','b', tau1-tau2)*Ivu('a','b', tau2-tau3) + \
    Ivu('b', 'c', t - tau1)*Ivu('b','b', tau1-tau2)*Ivu('a','b', tau2-tau3) + \
    Ivu('b', 'c', t - tau1)*Ivu('a','c', tau1-tau2)*Ivu('a','b', tau2-tau3)
````

````python
integrand = simplify(powsimp(expand(integrand)))
wl, ws = symbols(r'\omega_L,\omega_s')
integrand = integrand.subs({Symbol(r'E_a'): Symbol(r'\epsilon_a') + wl + ws, Symbol(r'E_b'): Symbol(r'\epsilon_b') + ws, \
    Symbol('E_c'): Symbol(r'\epsilon_c')})

display(integrand)
````

### Sympy Gotchas

```Lambda``` captures context. For example, ```Lamba((a,b), expr with a,b)``` actually doesn't do anything with the parameters if a,b are already defined in scope. Use builtin ```lambda``` instead.  

Also, ```.subs()``` checks the assumptions associated with the symbol when substitutions are made. ```Symbol('a') != Symbol('a', real=True)```. This is confusing.

Lambdify works with string representations when it maps to numerical version of the thing. It might make more sense to convert to an internal representation of the variables that can't have, for example, Latex representations.

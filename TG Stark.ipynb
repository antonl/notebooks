```python
basepath = '/Volumes/jogilvie/2D/Data/14-01-05'

def read_batch(name, batch):
    '''reads a filename/batch pair and returns an iterable that gives the data
    '''
    
    batch_path = path.join(basepath, '{name:s}-batch{batch:02d}'.format(name=name, batch=batch))
    
    #by assumption only read t1 = 0
    #t1 = np.genfromtxt(path.join(batch_path, 't1pos.txt'))
    
    t2 = np.genfromtxt(path.join(batch_path, 't2pos.txt'))[:, 1]
    
    for i, t in enumerate(t2):
        file_path = path.join(batch_path, 
            '{name:s}{t2:02d}-{t1:02d}-{repeat:02d}.spe'.format(name=name, batch=batch, t2=i, t1=0, repeat=0))
           
        data = SpeFile(file_path)
        data.header.datatype = 3 # Hack to fix incorrect data output format flag
    
        analog_path = '{name:s}{t2:02d}-{t1:02d}-{repeat:02d}-analog'.format(name=name, batch=batch, t2=i, t1=0, repeat=0)
        analog1 = np.genfromtxt(path.join(batch_path, analog_path + '1.txt')).squeeze()
        analog2 = np.genfromtxt(path.join(batch_path, analog_path + '2.txt')).squeeze()
        yield (np.asfarray(data.data.squeeze()), analog1[argwhere(analog1 > -50)], analog2[argwhere(analog2 > -50)], t)
```

```python
from scipy.interpolate import interp1d
def interpolate_to_freq(axis, data, n=200, dn=False):
    # assume axis is in nm from blue to red
    c = 299.792458 # nm/fs
    min_e = (c / axis[-1])
    max_e = (c / axis[0])
    frequency, step = np.linspace(min_e, max_e, n, retstep=True)
    tmp = interp1d(c/ axis[::-1], data[::-1])
    return (frequency, tmp(frequency)) if not dn else (frequency, tmp(frequency), step)
```

```python
import scipy.signal as _signal

def fourier_window(data, width=140, center=30, plot=False):
    n = width
    c = center
    filter_length = 200
    k = filter_length/2
    win = _signal.get_window(('kaiser', n), filter_length)
    win_shift = fft.fftshift(win)
    
    dc = np.fft.fft(data)
    ac = np.copy(dc)
    
    dc[:k] *= win_shift[:k]
    dc[-k:] *= win_shift[k:]
    dc[n:-n] = 0
    
    ac[c-k:c+k] *= win
    ac[:c-k] = 0
    ac[c+k:] = 0
    
    if plot:
        original = np.abs(np.fft.fft(data))
        
        plt.plot(win_shift[:k])
        grid()
        plt.plot(original[:c+k])
        scale = max(original[c-k:c+k])
        plt.plot(np.arange(filter_length) + c - k, win*scale)
        ylim([0, scale])
    return abs(np.fft.ifft(ac)), abs(np.fft.ifft(dc))
```

```python
# make lambda axis
center_wavelength = 530 # qt.nm
pixels = 1340
spectrometer_bandwidth = 140 # qt.nm
axis_lambda = np.linspace((center_wavelength - spectrometer_bandwidth/2), 
            (center_wavelength + spectrometer_bandwidth/2), 
            pixels)

maxt = 100
tg_fieldoff = np.empty((pixels, maxt))
tg_fieldon = np.empty((pixels, maxt))
t2_list = np.empty((maxt,))

fig, (ax1, ax2, ax3) = subplots(3, 1, sharex=True)

for i, (data, analog1, analog2, t2) in enumerate(it.islice(read_batch('FXN_77K_TG_Stark_3p5KV', 2), maxt)):
    # Do the synchronization boogie
    if i == 0:
        # first file, so keep all analog inputs
        pass
    else:
        # truncate the first three
        analog1 = analog1[3:]
        analog2 = analog2[3:]
    
    # truncate frames to smallest of analog, data
    if data.shape[0] > analog1.shape[0]:
        data = data[:analog1.shape[0]]
    elif data.shape[0] < analog1.shape[0]:
        analog1 = analog1[:data.shape[0]]
        analog2 = analog2[:data.shape[0]]
    
    #print data.shape, analog1.shape
    
    # In our case, the photodiode thing is useless, so ignore analog1
    
    # Find the maximum of the analog2 vector. It should correspond to the field on condition, and assume that the 
    # chopper and the electronics remain synchronized, but the camera frame correspondence is lost
    # Taking mod 4 of the index, you get the offset in camera frames. You can reconstruct everything from there
    # The sequence goes pumped stark, unpumped no stark, pumped no stark, unpumped stark
    offset = (analog2.argmax()) % 4
    
    mean_data = {
        'pump_stark':     mean(data[((offset)%4)::4, :], axis=0),
        'nopump_nostark': mean(data[((offset+1)%4)::4, :], axis=0),
        'pump_nostark':   mean(data[((offset+2)%4)::4, :], axis=0),
        'nopump_stark':   mean(data[((offset+3)%4)::4, :], axis=0)
    }
    
    # Fourier filter all of these
    for k in mean_data.keys():
        # interpolate data to frequency
        axis_f, filtered = interpolate_to_freq(axis_lambda, mean_data[k], n=axis_lambda.shape[0])
        ac, LO_estimate = fourier_window(filtered, center=125, width=80, plot=False)
        mean_data[k] = ac
    
    tg_fieldoff[:, i] = a = mean_data['nopump_nostark'] - mean_data['pump_nostark']
    tg_fieldon[:, i] = b = mean_data['nopump_stark'] - mean_data['pump_stark']
    
    
    ax1.plot(axis_f, a)
    ax2.plot(axis_f, b)
    ax3.plot(axis_f, a-b)
    
    
    t2_list[i] = t2
    xlabel('Frequency /  x 10^15 Hz')
    print i,

#savefig('tg_figs.png')
```

```python
import sys, time
try:
    from IPython.core.display import clear_output
    have_ipython = True
except ImportError:
    have_ipython = False

class ProgressBar:
    def animate_ipython(self, iter):
        try:
            clear_output()
        except Exception:
            # terminal IPython has no clear_output
            pass
        print '\r', self,
        sys.stdout.flush()
        self.update_iteration(iter + 1)

    def update_iteration(self, elapsed_iter):
        self.__update_amount((elapsed_iter / float(self.iterations)) * 100.0)
        self.prog_bar += '  %d of %s complete' % (elapsed_iter, self.iterations)

    def __update_amount(self, new_amount):
        percent_done = int(round((new_amount / 100.0) * 100.0))
        all_full = self.width - 2
        num_hashes = int(round((percent_done / 100.0) * all_full))
        self.prog_bar = '[' + self.fill_char * num_hashes + ' ' * (all_full - num_hashes) + ']'
        pct_place = (len(self.prog_bar) / 2) - len(str(percent_done))
        pct_string = '%d%%' % percent_done
        self.prog_bar = self.prog_bar[0:pct_place] + \
            (pct_string + self.prog_bar[pct_place + len(pct_string):])

    def __str__(self):
        return str(self.prog_bar)
    
    def __init__(self, iterations):
        self.iterations = iterations
        self.prog_bar = '[]'
        self.fill_char = '*'
        self.width = 40
        self.__update_amount(0)
        if have_ipython:
            self.animate = self.animate_ipython
        else:
            self.animate = self.animate_noipython
```

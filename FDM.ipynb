# Understanding Filter Diagonalization

Follow the cookbook.

1. Choose a small frequency window ($\omega_\text{min}$, $\omega_\text{max}$) for analysis of time signal $c_n$
2. Set up evenly spaced grid on the range
3. Evaluate $U^{(p)}$

```python
import itertools
import os
import subprocess
import cStringIO
from collections import namedtuple
from __future__ import print_function, division
%pylab inline --no-import
matplotlib.rcParams['figure.figsize'] = (16, 8)

import sys
sys.path.append('/Users/aloukian/Documents/fdm')
import _fdm as fdm
```

```python
defs = np.array([ # in units 1/s
#    a, f, g, phi
    [10, 4, -1.5j, 0],
    #[5, 15, -0.7j, 0],
    [15, 7, -3.01j, 0],
    ])

# Search for modes within frequencies freq in Hz
dt = 0.01

n = np.arange(100)

a_noise, p_noise =  0*np.random.random_sample(n.shape), 0*2*np.pi*np.random.random_sample(n.shape)
sig = np.sum([a*np.exp(-1.j*(g + 2*np.pi*f)*n*dt + phi*1.j) for a,f,g,phi in defs], axis=0) 
cn = sig + 0.5*a_noise*np.exp(1j*p_noise)
#plt.subplot(2,1,1)
#plt.plot(n*dt, np.imag(cn))
plt.plot(n*dt, np.real(sig))
```

```python
# Step 1

# canonical frequency will be positive, range 0 to 2 pi
fmin, fmax = 0, 10

diff = fmax - fmin
rho = cn.shape[0]*dt/2
#J = int(cn.shape[0]*dt*diff/2) + 2 # look for at least twods
J = 500
print("Expected spectral density is {}".format(rho))
print("Average inverse spacing is then {}".format(1/rho))
print("Expected number of spectral modes used in decomposition is {}".format(J))
print("By the way, frequencies are ambiguous outside the interval -{range:3.1f} to {range:3.1f}".format(range=1/(2*np.pi*dt)))
# Step 2
#d->z[i] = cexp(-I * TWOPI * (fmin + i * ((fmax - fmin) / (nf - 1))));

#wj = np2*np.pi*fmin*dt, 2*np.pi*fmax*dt, int(J)) 
wj = 2*np.pi*np.array([(fmin + x*(diff)/(J-1))for x in xrange(J)])
zj = np.exp(-1j*wj*dt)

plt.polar(np.angle(zj), np.abs(zj), marker='.')

poles = np.array([-1j*dt*(g + 2*np.pi*f) for a,f,g,phi in defs])
plt.polar(np.angle(np.exp(poles)), np.abs(np.exp(poles)), marker='o')
```

Planck's constant: $h = 4.13566751\times 10^{-15}\ \text{eV/Hz}$

Speed of light: $c = 299 792 458\ \text{m/s}$

Conversion: $E = hc/\lambda$

```python
def interpolate_ev(axis, data, n=200, dE=False):
    # assume axis is in nm from blue to red
    k = 4.13566751e-15 * 299792458 * 1e9
    min_e = (k / axis[-1])    
    max_e = (k / axis[0])
    energy,step = np.linspace(min_e, max_e, n, retstep=True)
    tmp = interp1d(k/ axis[::-1], data[::-1])
    return (energy, tmp(energy)) if not dE else (energy, tmp(energy), step)
```

```python
def fourier_window(data, n=140, plot=False):
    win = fft.fftshift(signal.get_window(('kaiser', 140), 200))
    dat_windowed = fft.fft(data)
    dat_windowed[0:100] *= win[0:100]
    dat_windowed[-100:] *= win[100:]
    dat_windowed[100:-100] = 0
    if plot:
        plot(win)
        plot(data)
    return abs(fft.ifft(dat_windowed))
```

```python
def interpolate_ev(axis, data, n=200, dE=False):
    # assume axis is in nm from blue to red
    k = 4.13566751e-15 * 299792458 * 1e9
    min_e = (k / axis[-1])    
    max_e = (k / axis[0])
    energy,step = np.linspace(min_e, max_e, n, retstep=True)
    tmp = interp1d(k/ axis[::-1], data[::-1])
    return (energy, tmp(energy)) if not dE else (energy, tmp(energy), step)
```

```python
basepath = '/Volumes/jogilvie/2D/Data/14-03-07/Stark/'
import tqdm
lamb_axis = wmap2(np.arange(calib_wavelengths.shape[0]), p0, p1, p2, p3)

def make_stark(paths):
    for path in tqdm.tqdm(paths):
        data = read_file(os.path.join(basepath, path))
        
        # skip 10 frames
        skip = 10
        even = mean(data[skip::2,:], axis=0)
        odd = mean(data[skip+1::2,:], axis=0)
        
        even_ev = interpolate_ev(lamb_axis, even, n=even.shape[0])
        odd_ev = interpolate_ev(lamb_axis, odd, n=odd.shape[0])
        axis = even_ev[0]
        yield (even, odd, axis, np.log10(fourier_window(even_ev[1])) - np.log10(fourier_window(odd_ev[1])), path)
```

```python
def prepare_even_odd(axis_lambda, filename):
    data = read_file(filename)
    data_e = data[0::2].mean(axis=0)
    data_o = data[1::2].mean(axis=0)

    axis_e, data_e= interpolate_ev(axis_lambda, data_e, n=axis_lambda.shape[0])
    axis_e, data_o = interpolate_ev(axis_lambda, data_o, n=axis_lambda.shape[0])
    return axis_e, data_e, data_o
```

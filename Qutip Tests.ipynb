## Constant field (time-independent Hamiltonian)

First, I wanted to a really simple system because I am stupid. I set up the "energies" in THz because they are divided by $\hbar$ during simulation. I suspect they really should be angular frequencies, because that would make more sense. 

If the coupling is time-independent, then the energy levels just shift. The peak shows up at the new energies. 

```python
#Ef = E.rescale('eV/m')
#q.radian =  q.UnitQuantity('radian', ((2*pi)*q.dimensionless), 'rad')

# hbar is assumed to be 1
w1 = -1*q.THz/2 *2*pi
w2 = 1*q.THz/2 *2*pi 
coupling = sqrt(1)*q.THz/2 *2*pi
Hs = float(w1.rescale('THz'))*qt.ket2dm(qt.basis(2, 0)) + float(w2.rescale('THz'))*qt.ket2dm(qt.basis(2, 1))
V = float(coupling.rescale('THz'))*qt.sigmax()
display(Hs)
display(V)
```

So the total Hamiltonian has off-diagonal elements, coupling is time-independent. Now calculate the dynamics. There is no collapse. The system is in a superposition of the ground and excited state.

```python
tlist, dt = np.linspace(0/(q.THz), 10/(q.THz), 1000, retstep=True) # In units of 1/THz
#tlist, dt = q.Quantity(tlist, '1/THz'), 1/rad*q.Quantity(dt, '1/THz')
res = qt.mesolve(Hs + V, (1*qt.basis(2, 0) + 1*qt.basis(2,1)).unit(), tlist.magnitude, [], [V, qt.ket2dm(qt.basis(2,0)), qt.ket2dm(qt.basis(2,1))])
```

```python
fig, (ax1, ax2) = subplots(2, 1, sharex=True)
ax1.plot(tlist.rescale('ps'), real(res.expect[1]))
ax1.plot(tlist.rescale('ps'), real(res.expect[2]))
ax1.set_title('Populations')
ax2.plot(tlist.rescale('ps'), real(res.expect[0]))
ax2.set_title('Polarization')
xlabel('time, ps')
#xlim([0, 1]);
```

Anyway, here's the Fourier transform of the spectrum. Notice the real and imaginary parts are identical. The phase is messed up, so we can take the absolute value and get the peak location.

```python
# This factor of 2*pi comes from FFT
freq = q.Quantity(fftshift(fftfreq(len(res.times), dt)), 'THz')*(2*pi)
ki = fftshift(fft.fft(res.expect[0]))
plot(freq, abs(ki))
#plot(freq, imag(ki))
#plot(fftshift(fftfreq(len(res.times), dt)), fftshift(imag(fft.fft(res.expect[0]))))
xlim([1, 20])
xlabel('Frequency (THz)')
#ylim([-0.5, 0.5])
```

Sure enough, the signal shows up at the difference between energies in the diagonalized Hamiltonian. 

Now, if we add a decay, we get a Lorenzian in the same location. The real and imaginary parts even look appropriate.

```python
tlist, dt = np.linspace(0, 10, 20000, retstep=True) # In units of 1/THz
tlist, dt = q.Quantity(tlist, '1/THz'), q.Quantity(dt, '1/THz')
res = qt.mesolve(Hs + V, (1*qt.basis(2, 0) + 0*qt.basis(2,1)).unit(), tlist.magnitude, [-sqrt(1)*qt.sigmaz()], [V])

f, (ax1, ax2) = subplots(2, 1)
ax1.plot(tlist.rescale('ps'), real(res.expect[0]))
ax1.set_xlabel('time, ps')
ax1.set_xlim([0, 10])

freq = q.Quantity(fftshift(fftfreq(len(res.times), dt)), 'THz')*(2*pi)
ki = fftshift(fft.fft(len(res.times)*res.expect[0]))
ax2.plot(freq, real(ki))
ax2.plot(freq, imag(ki))
ax2.set_xlim([700, 715])
#ax2.set_ylim([-0.5e9, 0.5e9])
#ax2.set_yscale('log')
ax2.set_xlabel('frequency, THz')
```

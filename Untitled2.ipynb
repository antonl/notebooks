```python
rate = 5.e6 # 5 MSa/s rate
length = 499e-6 # length of gaussian in ms
```

```python
n = np.arange(int(rate*length))
buf = len(n)/15
l4 = (len(n) - 2*buf)/4

wave = np.zeros((len(n), ))
wave[buf:buf+l4] = np.linspace(0, 1, num=l4)
wave[buf+l4:buf+3*l4] = 1
wave[buf+3*l4:buf+4*l4] = np.linspace(1, 0, num=l4)
        
gaussian = np.exp(-(n - len(n)/2)**2/(len(n)/25)**2)
gaus = partial(np.convolve, gaussian, mode='same')
swave = gaus(wave)
# normalize window
swave *= 1./swave.max()
waveform = np.asarray(swave*(2**15-1), dtype='>i2')
#self.verify_plot(self.waveform)
```

```python
plot(n*1/rate, swave)
```

```python
plot(np.diff(swave))
```

```python
q.pF = q.UnitQuantity('picofarad', definition=1e-12*q.farad, symbol='pF') 
```

```python
max_V = 10*q.kV
sample_period = 1./rate * q.s
total_c = 40*q.pF + 65*q.pF 
(total_c*np.max(np.diff(swave))*max_V/sample_period).rescale('mA')
```

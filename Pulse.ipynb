## Interference things, multiple pulses

Now, set up a Michelson interferometer and see what happens. As it turns out, just adding two pulses in the time domain doesn't give you interference. Looking at *intensity* does that. 

````python
# Put a pulse, and delay it. Recombine and see what happens
tau = sp.Symbol('tau', dummy=True) # Time delay var
a = fg1 # reuse our previous pulse
#a_delay = sp.Lambda(tau, a.subs({t : t - tau}))
#display(a_delay)

sigma = sp.re(a + a.subs({t : t - tau}))**2
nsigma = np.vectorize(sp.lambdify((t, tau), sigma, 'numpy'))
nt = np.linspace(-0.050, 0.050, 200)
ntau = linspace(-0.015, 0.015, 500)
#display(sigma(nt))
import itertools as it
intr = np.zeros(ntau.shape)
for i,j in it.izip(range(0, len(ntau)), ntau):
    s = nsigma(nt, j)
    #print i,j
    #plot(nt, s);
    intr[i] = np.average(s)

plot(ntau, intr)
````

*Finally!* This is what the interference scan should look like. I don't really have additional fringes, but, that's ok. 

Interestingly, if you scan overlap beyond the sampled time range, you get a large drop at the far ranges. This is because the average is calculated over a smaller number of samples that actually add up to a constant (I guess). 

````python
dt = np.linspace(0.000, 0.010, 400)
X,Y = np.meshgrid(nt, dt)

res = nsigma(X, Y) 
imshow(res, vmin=0., vmax=2500., extent=[np.min(nt), np.max(nt), np.min(dt), np.max(dt)], origin='lowe
````

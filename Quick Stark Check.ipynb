```python
def interpolate_fs(axis, data):
    # assume axis is in nm from blue to red
    k = ut.speed_of_light.to('nm/ps').magnitude
    min_f = (k / axis[-1])    
    max_f = (k / axis[0])
    freq,step = np.linspace(min_f, max_f, data.shape[0], retstep=True)
        
    tmp = interp1d(k/ axis[::-1], data[::-1])
    return (freq, tmp(freq), step)
```

```python
def fourier_window(data, n=200, plt=False, k=0.1):
    edge_window = get_window(('kaiser', k), len(data))
    win = fftshift(get_window(('kaiser', 140), n))
    fdat = fft(data*edge_window)
    dat_windowed = fdat.copy()
    dat_windowed[0:n//2] *= win[0:n//2]
    dat_windowed[-n//2:] *= win[n//2:]
    dat_windowed[n//2:-n//2] = 0
    
    filtered = abs(ifft(dat_windowed))
    if plt:
        subplot(211)
        scale = max(edge_window)
        plot(data)
        plot(scale*edge_window)
        plot(data*edge_window)
        plot(filtered)
        subplot(212)
        scale = max(abs(fdat[5:-5]))
        plot(win[:n//2]*scale)
        plot(abs(dat_windowed))
        plot(abs(fdat))
        ylim([0, scale])
        tight_layout()
    return filtered

#fourier_window(s3p5_data[::2].mean(axis=0), n=1000, plt=True)
```

```python
def fix_baseline(data, range):
    mean = data[range].mean()
    return data-mean
```

```python
phases = [1, -1, 1, -1, -1, 1, 1, 1, 1]
#sign = [1, 1, 1, 1, 1, 1, 1, 1, 1]
diff_list = []
data_list = [d1d2_baseline, d1d2_0p51, d1d2_0p75, d1d2_1p01, d1d2_1p51, d1d2_2p01, d1d2_3p01, d1d2_4p01, d1d2_5p01]
peaks_idx = [890, 995, 1064]
voltages = array([0., 0.5, 0.75, 1.0, 1.5, 2.0, 3.0, 4.0, 5.0])
fig = figure()
ax1 = subplot2grid((2, 2), (0, 0), colspan=2)
ax2 = subplot2grid((2, 2), (1, 0))
ax3 = subplot2grid((2, 2), (1, 1))


def filterer(x):
    if x[0] in [0, 1, 3, 5, 8]: 
        return True
    else:
        return False
    
for sign, (even, odd) in izip(phases, analyze_stark(data_list)):
    #fourier_window(even, n=250, k=5, plt=True)
    res = np.log10(fourier_window(even, n=250, k=0.1)) - np.log10(fourier_window(odd, n=250, k=0.1))
    #f, res_int, df = interpolate_fs(wl, res)
    res = fix_baseline(res, slice(0, 500))
    diff_list.append(sign*res[::-1])
    ax1.plot(wl, sign*res[::-1])
    #print peaks
    ax2.plot(wl, sign*res[::-1])

ax2.vlines(wl[peaks_idx], np.min(res), max(res))
ax2.set_xlim(650, 710)

scans = np.rollaxis(np.array([d[peaks_idx] for d in diff_list]), 1)
ssign = [1, 1, -1]
fn = lambda p, x: p[2]*x**2 + p[1]*x + p[0] 
err = lambda p, x, y: y - fn(p, x)
fit_v = linspace(0, 5, 100)
    
for i, s in ifilter(filterer, izip(xrange(len(peaks_idx)), ssign)):
    ax3.plot(voltages, s*scans[i, :], label='{:.1f}'.format(wl[peaks_idx[i]]), marker='o')
    res = leastsq(err, [0.1, 0.1, 0.1], args=(voltages, s*scans[i, :]), full_output=True)
    ax3.plot(fit_v, fn(res[0], fit_v), 'r-.')
    
ax3.set_xlim(-0.1, 6)

fmt = ScalarFormatter(useOffset=False, useMathText=False)
fmt.set_scientific
ax2.xaxis.set_major_locator(AutoLocator())
#ax2.xaxis.set_minor_locator(MultipleLocator(2))
ax2.yaxis.set_major_locator(AutoLocator())
ax2.yaxis.set_major_formatter(ScalarFormatter(useOffset=False, useMathText=False))
#ax2.yaxis.set_minor_locator(AutoLocator())
ax2.grid(which='major', axis='x', linewidth=1, linestyle='-.', color='0.75', alpha=0.8)
#ax2.grid(which='minor', axis='x', linewidth=0.25, linestyle='--', color='0.75')
ax2.grid(which='major', axis='y', linewidth=1, linestyle='-.', color='0.75', alpha=0.8)
#ax2.grid(which='minor', axis='y', linewidth=0.25, linestyle='--', color='0.75')
```

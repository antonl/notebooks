FFT of length $N$ definitions:
$$\begin{eqnarray}
\text{fft:} \quad f(k) & = & \quad \sum_{n = 0}^{N - 1} e^{-2\pi i \frac{k n}{N}} f(n) \\
\text{ifft:} \quad f(n) & = & \frac{1}{N} \sum_{n = 0}^{N - 1} e^{2\pi i \frac{k n}{N}} f(k) \\
\end{eqnarray}$$


Conventions: 

- `dt`: time step for discretizing time
- `f`: complex frequency in units of `1/s`
- `w`: angular frequency normalized to the time step

```python
N = 4096*4
dt = 1e-4
n = arange(N) 
f = 100. - 1j*5
a = 5+1j
# Generate signal
signal = a*exp(-1j*2*pi*f*dt*n)
```

```python
fsignal =  2*pi*dt*fftshift(fft.fft(signal))
df = fftshift(fftfreq(N, dt))
df_an = arange(-150, -50, 0.1)
analytic = a*1./(imag(-f) + 1j*(df_an - real(-f)))
```

```python
fig, (ax1, ax2) = subplots(2, 1)
ax1.plot(n*dt, real(signal))
ax1.set_xlim([0, 0.2])
ax2.plot(df, real(fsignal), df_an, real(analytic))
ax2.set_xlim([-150, -50]);
```

The waveform generator is on the fritz, so we had to come up with a new way for generating the driving waveform. Currently, the plan is to use this waveform, where blue is signal and red is laser pulse.

```python
t = np.linspace(0, 2*T, 1000)
T = 1./250. # Hz
#k = 0.5

waveform = np.sin(2*np.pi/T*t)
waveform[np.argwhere(waveform < 0)] = 0
#plt.plot(t, np.sin(t))
plt.plot(t/T, waveform)
plt.vlines([0.25, 0.75, 1.25, 1.75], 0, 1, 'r') 
plt.xlabel('t/T')
```

We are trying to make this from a square wave. The Fourier expansion for a square wave with a variable duty cycle $k \in [0,1]$ is 

$$
\begin{equation}
F(t) \approx k T - \frac{4}{T} \sum_{n = 1, 3, 5} ^\infty \frac{1}{n}(\cos(2 \pi n k) - 1) \sin \left(\frac{2 \pi}{T} n t\right)
\end{equation}
$$

```python
k = 0.5
N = 1
signal = np.zeros_like(t)
signal += (k*T)/2
for i in xrange(N):
    n = 2*i + 1
    #print n
    signal += -1./T*1./n*(np.cos(2*np.pi*n*k) - 1)*np.sin(2*np.pi/T*n*t)

plt.plot(t/T, signal)
```

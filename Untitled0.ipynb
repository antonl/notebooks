```python
path = '/Volumes/jogilvie/2D/Anton/data_13Feb2014.mat'
```

```python
import scipy.io as io
import scipy.interpolate
```

```python
thing = io.loadmat(path)
```

```python
t2,angfreq,data = -thing['t2'][:, 1], thing['vWe'].squeeze(), thing['Stark_TG_sub'].T
```

```python
t2_plot = t2 * 0.001 # convert to ps
## interpolate to a nicer grid
stark_spline = scipy.interpolate.interp2d(angfreq, t2_plot, data)

min_t2_image = -0.2 # in ps
max_t2_image = 1 # in ps

t2_image, dt2_image = np.linspace(min_t2_image, max_t2_image, 500, retstep=True)
stark_image = stark_spline(angfreq, t2_image)

figure(figsize=(8,5))
imshow(stark_image, aspect='auto', interpolation='nearest',
       extent=(angfreq.min()-0.5*(angfreq[2]-angfreq[1]), angfreq.max()+0.5*(angfreq[2]-angfreq[1]), 
       t2_image.max()+0.5*dt2_image, t2_image.min()-0.5*dt2_image))
xlabel('Angular frequency (1/fs)')
ylabel('time (ps)')
title('Stark TG data on FXN (absolute value)')
colorbar()
#savefig('StarkTG.pdf')
savefig('StarkTG.png')
#plt.savefig('StarkTG.svg')
```

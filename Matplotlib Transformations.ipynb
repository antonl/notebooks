````python
def gaussian1d(A, w, t):    
    c = 4*log(2)
    return sqrt(c/np.pi)*A/w*np.exp(-c*t*t/(w*w))

def pulse(t, τ, φ=0.):
    '''create a pulse of the amplitude-carrier form :math:`A cos(ω(t-τ) + φ)`,
    with A being a Gaussian
    '''
    t0 = t-τ
    return gaussian1d(1., 20, t0)*np.cos(2*np.pi*(t0) + φ)
````

## Data

So matplotlib has several coordinate transformation objects that are used to convert data in userland coordinates (t, x) to display coordinates in a backend. The one most often used is the data transform. It is accessed through an ```Axes``` object, and converts points in the (t, x) coordinates straight to display coordinates. It also has an ```.inverted()``` method that returns the inverse of the transformation. These objects are used to place text at certain locations in relation to the data.

It seems like it is possible to use the transform to draw points outside of the current view limits. This might draw things outside of the axes I guess?

## Axes

The next useful transform is the Axes space transform. The ```Axes``` coordinate space has the origin at the bottom-left and grows until the top-right corner, where it is (1.0, 1.0). You can get the transform object by using the ```Axes.transAxes``` property.

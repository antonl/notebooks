## Calibrate spectrum

Given a calibration spectrum of a mercury-argon lamp, fit a polynomial to it and determine the wavelengths. This is a very manual process.

````python
data = np.squeeze(sf.data).mean(axis=-1)
data -= np.min(data)

res = argrelmax(data, order=10)[0]
select = data[res] > 0.2*np.max(data)
````

````python
offsets = list(zip(res[select], data[res[select]]))

ax = subplot(111)
ax.plot(data)
stars = matplotlib.collections.StarPolygonCollection(numsides=5, sizes=(50,), color=('r',), offsets=offsets, transOffset=ax.transData)
ax.add_collection(stars)
grid()
````

````python
data = np.squeeze(sf.data).mean(axis=-1)

fig, (ax1, ax2) = pyplot.subplots(2, 1)

# baseline subtraction
px = np.arange(data.shape[0])
pv = np.polyfit(px, data, 2)
data -= np.polyval(pv, px)

# peak fitting
res = argrelmax(data, order=10)[0]
select = data[res] > 0.2*np.max(data)

# set peaks
peaks = np.array([706.722, 696.543, 626.3688, 625.1348, 593.4566])

# plotting

ax1.plot(wl, data)

# known lines and their orders
norders = 3
color = 'gck'
lines = list(it.chain.from_iterable([i*nudie.spectrometer.hg_ar_lines for i in range(1, norders)]))
colors = list(it.chain.from_iterable([it.repeat(color[i], len(nudie.spectrometer.hg_ar_lines)) for i in range(1, norders)]))
offsets = list(zip(lines, it.repeat(-100)))

stars = matplotlib.collections.StarPolygonCollection(5, sizes=(10,), color=colors, offsets=offsets, transOffset=ax1.transData)
ax1.add_collection(stars)

corr = np.polyfit(res[select], peaks, 2)
wl = np.polyval(corr, px)

#for lines, c in lines_list:
#    ax1.vlines(lines, -0.3*np.max(data), 0, c, linewidth=2)

#ax1.vlines(np.polyval(corr, res[select]), 0, 0.3*np.max(data), 'r')
ax1.set_ylabel('Intensity / AU')
ax1.set_xlabel('Wavelength / nm')
ax1.set_xlim(min(wl), max(wl))
ax1.grid()
ax2.text(0.01, 0.05, r'$y = {:.3g} \cdot x^2 + {:.3g} \cdot x + {:.3g}$'.format(*corr), transform=ax2.transAxes, fontsize=16)
ax2.plot(px, np.polyval(corr, px))
ax2.set_ylabel('Wavelength / nm')
ax2.set_xlabel('Pixel');
````

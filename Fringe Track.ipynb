```python
import sys
sys.path.append('./GitHub/pyDAQmx/')
from itertools import ifilter
from daqmx import NIDAQmx, Task, AnalogInputVoltage, SampleClock
from daqmx.units import Volts
```

```python
s = NIDAQmx()
print s.version

print 'Devices:'
for d in s.devices: print '\t> ' + str(d)

a = Task('fringe_track')
a.add_channel(AnalogInputVoltage, s.inputs()[0], 0, 0.2, Volts, name='photodiode_1')
a.add_channel(AnalogInputVoltage, s.inputs()[1], 0, 0.2, Volts, name='photodiode_2')
print 'Tasks: '

for i in ifilter(lambda x: x() is not None, s._tasks_by_name.valuerefs()):
    print '\t> ' + str(i())
```

```python
a.set_timing(SampleClock, rate=1024*10, n_per_channel=1024*1000) 
```

```python
import time
a.start()

data = a.read_as('AnalogF64', buf_size =1024*1000, blocking=True)
print 'mean 1: {:4f}, mean 2: {:4f}'.format(numpy.mean(data.data[0::2]), numpy.mean(data.data[1::2]))
a.stop()
    
```

```python
from weakref import proxy
import numpy

class TaskReader(object):
    def __init__(self, task, buf_size, *args, **kwargs):
        self._step_callback = None
        self._task = proxy(task)
        self._buf_size = buf_size
        self._nchannels = len(task.channels)
        #TODO: get ability to query the way multiple tasks get converted to ndarray
        #self._group_by = task.samples_group_by
        
    def start(self):
        self._task.set_nsamples_callback(self._call, self._buf_size)

        print 'started task'
        self._task.start()
    
    def stop(self):
        try:
            self._task.stop()
        except RuntimeError as e:
            print e
    
    def add_callback(self, func):
        '''register function to call when source updates
        '''
        self._step_callback = func
        
    def remove_callback(self, func):
        self._step_callback = None
        try:
            self._task.stop()
        except RuntimeError as e:
            print e
        
    def _call(self, handle, eventType, nsamples, data):
        if self._step_callback is not None:
            data = self._task.read_as('AnalogF64', buf_size=nsamples, blocking=False)
            self._step_callback(data.data) # pass data interleaved
        return 0
    
from matplotlib.lines import Line2D
from matplotlib.animation import Animation
import numpy
    
class CallbackAnimation(Animation):
    def __init__(self, fig, event_source, buf_size, ylim=(0, 1), nchannels=1, skip=5, *args, **kwargs):
        self._buf_size = buf_size
        self._ylim = ylim
        self._nchannels = nchannels
        self._skip = skip
        super(CallbackAnimation, self).__init__(fig, event_source, blit=False, *args, **kwargs)
        
    def _step(self, data):
        self._draw_next_frame(data, self._blit)
        return True
    
    def new_saved_frame_seq(self):
        pass
    def new_frame_seq(self):
        pass
    
    def _init_draw(self):
        ax = self._fig.add_subplot(111)
        ax.set_xlim(0, self._buf_size)
        ax.set_ylim(self._ylim)
        self._lines = [Line2D(numpy.zeros((self._buf_size,)), numpy.zeros((self._buf_size,)), linewidth=1.) for x in xrange(self._nchannels)]
        for i in self._lines:
            print 'added line'
            ax.add_line(i)
    
    def _draw_frame(self, data):
        print 'drawing frame'
        x_data = numpy.zeros((len(data)/self._nchannels,))
        for i in xrange(self._nchannels):
            self._lines[i].set_xdata(x_data)
            self._lines[i].set_ydata(data[i::self._nchannels])
    
    def save(self):
        pass
```

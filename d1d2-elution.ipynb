# 12/08/2014 Concentrating D1D2 Stark Sample

````python
dp = mount_point / 'UV-Vis spectra'
datas = []
for x in dp.glob('d1d2-* (2014 Dec 08).csv'):
    with x.open('rb') as f:
        datas.append((x.stem, np.genfromtxt(f, delimiter=b',', skip_header=2, unpack=True)))
````

````python
for name, (wl, data) in datas:    
    wl750 = argmin(abs(wl - 750))
    wl416, wl435 = argmin(abs(wl - 416)), argmin(abs(wl - 435))
    #data -= data[wl750]
    plot(wl, data, label='{:s} {:1.2f}'.format(name, float(data[wl416]/data[wl435])))
axvline(wl[wl416], 0, 1.6)
axvline(wl[wl435], 0, 1.6)
axvline(wl[wl750], 0, 1.6)
xlabel('Wavelength / nm')
ylabel('Absportion / cm⁻¹')
legend()
grid()
wl677 = argmin(abs(wl - 677))
print("OD @ 677 nm", data[wl677])
#savefig('d1d2-nov10.png')
````

# 11/07/2014 D1D2 Prep

````python
dp = mount_point / 'UV-Vis spectra'
datas = []
for x in dp.glob('d1d2-batch* (2014 Nov 07).csv'):
    with x.open('rb') as f:
        datas.append((x.stem, np.genfromtxt(f, delimiter=b',', skip_header=2, unpack=True)))
````

````python
for name, (wl, data) in datas:    
    wl750 = argmin(abs(wl - 750))
    wl416, wl435 = argmin(abs(wl - 416)), argmin(abs(wl - 435))
    #data -= data[wl750]
    plot(wl, data, label='{:s} {:1.2f}'.format(name, float(data[wl416]/data[wl435])))
axvline(wl[wl416], 0, 1.6)
axvline(wl[wl435], 0, 1.6)
axvline(wl[wl750], 0, 1.6)
xlabel('Wavelength / nm')
ylabel('Absportion / cm⁻¹')
legend()
grid()
savefig('d1d2-nov07.png')
````

Clearly my treatment with Triton-X was insufficient. The ratio indicates some superposition of CP47-RC and RC. There probably isn't a significant amount of contamination with free RC. The proper ratios for future reference is ```CP47-RC: 0.90, RC: 1.20, CP47: 0.75```.

````python
with (dp / 'd1d2-preelution(2014 Nov 07).csv').open('rb') as f:
    name, d = ('preelution', np.genfromtxt(f, delimiter=b',', skip_header=2, unpack=True))

wl, data = d[0], d[1]

wl750 = argmin(abs(wl - 750))
wl416, wl435 = argmin(abs(wl - 416)), argmin(abs(wl - 435))

data = data - data[wl750]

plot(wl, data, label='{:s} {:1.2f}'.format(name, float(squeeze(data[wl416]/data[wl435]))))

axvline(675.5, 0, 1.6)
axvline(wl[wl416], 0, 1.6)
axvline(wl[wl435], 0, 1.6)
xlabel('Wavelength / nm')
ylabel('Absportion / cm⁻¹')
legend()
grid()
````

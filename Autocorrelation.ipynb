# Simulating Fourier Transform Spectroscopy

A Gaussian function parameterized by its full width at half maximum $w$ and amplitude $A$ is given by

$$g(x; A, w) = \sqrt{\frac{4 \ln 2}{\pi}} \frac{A}{w} \exp \left (- 4 \ln 2 \frac{x^2}{w^2}\right)$$

If I was clever, I would nondimensionalize the function to measure everything in FWHMs.

```python
def gaussian1d(A, w, t):    
    c = 4*log(2)
    return sqrt(c/np.pi)*A/w*np.exp(-c*t*t/(w*w))

def pulse(t, τ, φ=0.):
    '''create a pulse of the amplitude-carrier form :math:`A cos(ω(t-τ) + φ)`,
    with A being a Gaussian
    '''
    t0 = t-τ
    return gaussian1d(1., 20, t0)*np.cos(2*np.pi*(t0) + φ)
```

```python
t = np.linspace(-60, 60, 2000)
plot(t, pulse(t, 0))
grid()
```

```python
dτ = 0.1
t, τ  = np.arange(-200, 200, 0.1), np.arange(-100, 100, dτ)
tv, τv = np.meshgrid(t, τ, indexing='ij')

autocorrelation = np.sum((pulse(tv, τv) + pulse(tv, 0))**2, axis=0)
plot(τ, autocorrelation)
grid()
```

```python
powerspectrum = np.fft.fftshift(np.fft.fft(autocorrelation - autocorrelation.mean()))
freq = np.fft.fftshift(np.fft.fftfreq(τ.shape[0], dτ))
plot(freq, 2*np.abs(powerspectrum))
xlim(0, 2)
grid()
```

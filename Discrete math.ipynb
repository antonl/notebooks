```python
from __future__ import division
pi = np.pi

%reload_ext sympy.interactive.ipythonprinting
%config InlineBackend.figure_format = 'svg'

def get_lines(pts):
    ''' Returns vlines corresponding to points in pts '''
    ymin, ymax = [], []
    for p in pts:
        if p >= 0: ymin.append(0), ymax.append(p)
        else: ymin.append(p), ymax.append(0)
    return ymin, ymax

def filter_dict(vals, acceptable=[]):
    ok = {}
    
    for k, v in vals.iteritems():
        if k in acceptable:
            ok.update({k : v})
    return ok

def mplot(x, y, **kwargs):
    ''' Make a plot that looks like the one in the book'''
    opt = {'color': 'black', 'linestyle':'solid', 'marker': 'o'}
    opt.update(kwargs)
    #plot(a1, a2, **opt)
    a,b = get_lines(y)
    
    scatter(x, y, **filter_dict(opt,['s', 'c', 'marker', 'cmap', 'norm', 'vmin', 
        'vmax', 'alpha', 'linewidths', 'faceted', 'verts', 'color']))
    
    vlines(x, a, b, **filter_dict(opt, ['ymin', 'ymax', 'color', 'linestyles']))
    
    b1, b2 = np.min(y), np.max(y)
    # make amplitude checking work right
    ylim(-1.1, 1.1) # hack for now

def dcos(A=1, f=2, theta=0):
    return np.vectorize(lambda n: A*np.cos(2*np.pi*f*n + theta))
```

## Discrete time

Use these definitions:

- $\omega \equiv 2\pi f$ where $f$ is in cycles (radians) per sample
- Discrete time sinusoid is then $ x(n) = A \cos(2\pi f n + \theta)$ with $-\infty \lt n \lt \infty$

## Properties of discrete time sinusoids

In contrast to continuous-time signals, a discrete sinusoid is:

1. periodic only if its frequency $f$ is a rational number
2. discrete-time sinusoids are identical if their frequencies $\omega$ are separated by an integer multiple of $2\pi$. The flip side of this is that signals with frequencies in the interval $-\pi \leq \omega \leq \pi$ or $-\frac{1}{2} \leq f \leq \frac{1}{2}$ are unique. The signals outside this range are called *aliases*


```python
# Aliasing
n = np.arange(-25, 25)
ex3 = dcos(A=1, f=0.1, theta=0)
ex4 = dcos(A=1, f=1.1, theta=0)

mplot(n, ex4(n), color='red')
mplot(n, ex3(n), color='blue', marker='.')
```

```python
ex2 = dcos(A=1, f=0.1, theta=0.1)
ex3 = dcos(A=1, f=pi/35, theta=0.1)

n = np.arange(-15, 15)
subplot(121)
mplot(n, ex2(n), color='red')
title('Periodic for rational $f$')

subplot(122)
n = np.arange(0, 30)
mplot(n, ex3(n), color='blue')
title('Aperiodic for irrational $f$');
```

See, the sequence is periodic if $f$ can be written as a rational number.

## Link to continuous time and sampling

The link between analog and discrete signals is the sampling function. This is given by $x(n) = \mathcal X(kT) = \mathcal X(k/F_\text{S})$, where $\mathcal{X}$ is the analog signal and $F_\text{S}$ is the sampling frequency. 

Using this definition, $\mathcal X(kT) \equiv x(n) = A \cos(2\pi F kT + \theta) = A \cos\left(2\pi n \frac{F}{F_\text{S}}\right)$. So, $f$ is called the normalized frequency and is defined as $f \equiv \frac{F}{F_\text{S}}$.

This definition yields the Niquist criterion, $-\frac{F_\text{S}}{2} \leq F \leq \frac{F_\text{S}}{2}$

## Decomposition of discrete time signal

It is convenient to use the linear property of a system by tracking the behavior of a basis function through the system and using this information to get the behavior of an arbitrary input signal. A commonly used basis is the impulse basis: $\{\delta(n-k)\ \forall k \in \mathcal Z\}$.

An arbitrary signal is represented as $x(n) = \sum_{k = -\infty}^{\infty} x(k)\,\delta(n-k)$. Now, the response of a linear system can be represented as 

$$y(n) = \mathcal T[x(n)] = \mathcal T \left[\sum_{k = -\infty}^{\infty} x(k)\delta(n-k)\right] = \sum_{k = -\infty}^{\infty} x(k) \mathcal T [\delta(n-k)]$$

If we have time-invariance, we can use the fact that if $h(n) \equiv \mathcal T[\delta (n)]$  then $h(n-k) = \mathcal T[\delta(n-k)]$ to express the above formula as a convolution:

$$y(n) = \sum_{k = -\infty}^{\infty} x(k)\ h(n-k)$$

This implies that the LTI system is completely characterized by its impulse response function.

### Properties of convolution

- Commutative $x(n) * h(n) = h(n) * x(n)$. Comes from substitution of $j = n-k$ in sum. 
- Associative $[x(n) * h_1(n)] * h_2(n) = x(n)*[h_1(n) * h_2(n)]$. 
- Distributive $x(n)*[h_1(n) + h_2(n)] = x(n)*h_1(n) + x(n) * h_2(n)$

### Causality

For a causal system, the impulse response due to future values has to be zero, or $h(n) = 0\ \forall n \lt 0$. Thus, can represent causal signals in these ways

$$y(n) = \sum_{k = 0}^\infty h(k) x(n-k) = \sum_{k=-\infty}^n x(k) h(n-k)$$

### Stability

A LTI system is stable if the impulse response is absolutely summable. 

## Fourier Series and so on

### Continuous time
To have a Fourier series expansion, the function of interest should have a norm in $L_2$. In that case, the series converges in the mean (least-squares sense). However, if the integral $\frac{1}{T_\text{period}}\int f(x)\, dx$ exists, then the expansion converges absolutely. Most physical signals work. 

Basically, you use Fourier's trick to get the coefficients of the expansion.

As $T_\text{period}$ increases, the "spacing" between points decreases (proportional to the inverse of the period). Increasing the width of the signal decreases the breadth of the series coefficients.

With this in mind, if we have a finite aperiodic signal and let the period tend to infinity, $x(t) \equiv \displaystyle\lim_{T_\text{period} \rightarrow \infty} x_\text{periodic}(t)$, in Fourier space the spacing between terms (lines) becomes smaller and smaller until its spectrum is continuous. This means that this spectrum will be the envelope of the spectrum obtained by repeating this aperiodic signal with a certain period. 

We can obtain this directly by applying the Fourier transform, which implies that we let $T_\text{period} = \infty$. 

Further, the Fourier series of the repeated periodic signal is the Fourier transform of the aperiodic signal sampled at regular intervals:

$$c_k = \frac{1}{T_\text{period}} X(kF_0) = \frac{1}{T_\text{period}}X\left(\frac{k}{T_\text{period}}\right)$$

This supports the interpretation that it is an envelope.

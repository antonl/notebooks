```python
# Functions taken from http://nicky.vanforeest.com/misc/fitEllipse/fitEllipse.html
# and slightly corrected
def fit_ellipse(x,y):
    x = x[:,np.newaxis]
    y = y[:,np.newaxis]
    D =  np.hstack((x*x, x*y, y*y, x, y, np.ones_like(x)))
    S = np.dot(D.T,D)
    C = np.zeros([6,6])
    C[0,2] = C[2,0] = 2; C[1,1] = -1
    E, V =  np.eig(np.dot(inv(S), C))
    n = np.argmax(np.abs(E))
    a = V[:,n]
    return a

def ellipse_center(a):
    b,c,d,f,g,a = a[1]/2, a[2], a[3]/2, a[4]/2, a[5], a[0]
    num = b*b-a*c
    x0=(c*d-b*f)/num
    y0=(a*f-b*d)/num
    return np.array([x0,y0])

def ellipse_angle_of_rotation( a ):
    b,c,d,f,g,a = a[1]/2, a[2], a[3]/2, a[4]/2, a[5], a[0]
    
    if np.abs(b) < 1e-6:
        if a > c: return np.pi/2
        else: return 0.
    else:
        if a > c: return np.pi/2 + 0.5*np.arctan2(2*b, a-c)
        else: return 0.5*np.arctan2(2*b, a-c)

def ellipse_axis_length( a ):
    b,c,d,f,g,a = a[1]/2, a[2], a[3]/2, a[4]/2, a[5], a[0]
    up = 2*(a*f*f+c*d*d+g*b*b-2*b*d*f-a*c*g)
    down1=(b*b-a*c)*( (c-a)*np.sqrt(1+4*b*b/((a-c)*(a-c)))-(c+a))
    down2=(b*b-a*c)*( (a-c)*np.sqrt(1+4*b*b/((a-c)*(a-c)))-(c+a))
    res1=np.sqrt(up/down1)
    res2=np.sqrt(up/down2)
    return np.array([res1, res2])
```

# Fitting an Ellipse for Position Calibration

I have a two signals that have some fixed (or slowly varying) phase relationship with each other. I want to determine what that fixed relationship is despite gain mismatch, differing offsets and additive noise.

## Synthetic data

Make an interferometer. The two beams are $E_1(t) = E_1 \sin(\omega t)$, $E_2(t) = E_2 \sin(\omega t + \delta)$. Recombine to obtain fringes: $I(t) = (E_1(t) + E_2(t))^2$ and $Q(t) = I(t)\cos(\delta)$. This is time fringe quadrature instead of spacial fringes. After integrating over the optical period you get

$$
\begin{eqnarray}
I(t) &=& \frac{A_{1} E_{1}^{2}}{2} + A_{1} E_{1} E_{2} \cos{\left (\omega \tau \right )} + \frac{A_{1} E_{2}^{2}}{2} + K_{1} \\
Q(t) &=& \frac{A_{2} E_{1}^{2}}{2} + A_{2} E_{1} E_{2} \cos{\left (\omega \tau + \delta \right )} + \frac{A_{2} E_{2}^{2}}{2} + K_{1} + K_{2}
\end{eqnarray}
$$


Here's some synthetic data. My two channels are represented by $I$ and $Q$. Initially, I assume that $A$, $B$, $C$, and $D$ are time-independent.


```python
# Define form and obtain interference terms
E1,E2,K1,K2,w,phi,tau,A1,A2,delta = sp.symbols('E1,E2,K1,K2,omega,phi,tau,A1,A2,delta', constant=True, real=True, positive=True)
t = sp.symbols('t', real=True)
p = sp.Wild('p')

I = A1*(E1*sp.sin(w*t) + E2*sp.sin(w*(t+tau)))**2 + K1
Q = I.subs({A1: A2, tau: tau+delta/w}) + K2

Q = Q.expand(trig=True)
I = I.expand(trig=True)
I,Q
```

```python
# Integrate over optical period
tmp = sp.piecewise_fold(((1/(2*sp.pi/w))*I).expand(trig=True).integrate((t, 0, 2*sp.pi/w)))
I = tmp.trigsimp()
tmp = sp.piecewise_fold(((1/(2*sp.pi/w))*Q).expand(trig=True).integrate((t, 0, 2*sp.pi/w)))
Q = tmp.trigsimp()
I,Q
```

So the above should be my $x,y$ coordinates. Any ellipse can be represented in the general quadratic curve form, 

$$a x^2 + 2bxy + cy^2 + 2dx + 2fy + g = 0$$

The ellipse has the following constraints: 

$$
\begin{eqnarray}
\Delta &=& \begin{vmatrix} a & b & d \\ b & c & f \\ d & f & g \end{vmatrix} \ne 0\\
J &=& \begin{vmatrix} a & b \\ b & c \end{vmatrix} \gt 0 \\
I &=& a + c \\
\Delta/I &\lt& 0
\end{eqnarray}
$$

Originally, I thought that since I know $(I,Q)$ trace out an ellipse, by taking $I^2 + IQ/2 + Q^2 + I/2 + Q/2$ and grouping terms by $\cos(\omega \tau)$ and $\cos(\omega \tau + \phi)$ terms, I should be able to identify the coefficients as $a,b,c,d,f,-g$. This doesn't work. Instead I have to take my $I,Q$ terms as my $x,y$ and substitute into the formal equation. Then I solve for the coefficients $a,b,\dots,g$

Apparently this doesn't work. Instead, try substituting into the equation and then solve for parameters.


```python
# Substitute into the quadratic implicit form
x,y = sp.symbols('x,y', dummy=True)
a,b,c,d,f,g = sp.symbols('a b c d f g')

Fxy = (a*x**2 + 2*b*x*y + c*y**2 + 2*d*x + 2*f*y + g).subs({x: I, y: Q}).expand(trig=True)
Fxy = Fxy.replace(sp.cos(p)**2, (1 + sp.cos(2*p))/2).replace(sp.sin(p)**2, (1 - sp.cos(2*p))/2).expand()
#display(Fxy)
```

```python
# Obtain coefficients and group terms
A = Fxy.coeff(sp.sin(2*w*tau))
B = Fxy.coeff(sp.cos(2*w*tau))
C = Fxy.coeff(sp.cos(w*tau)*sp.sin(w*tau))

D = Fxy.coeff(sp.sin(w*tau)).coeff(sp.cos(w*tau), n=0)
F = Fxy.coeff(sp.cos(w*tau)).coeff(sp.sin(w*tau), n=0)
G = Fxy.coeff(sp.cos(w*tau), n=0).coeff(sp.sin(w*tau), n=0).coeff(sp.cos(2*w*tau), n=0).coeff(sp.sin(2*w*tau), n=0)
```

We are free to choose the $g$ parameter because you can scale the implicit form by a constant without changing the ellipse it represents. Set $g \equiv 1$.

```python
sol = sp.solve([x for x in [A,B,C,D,F,G]], (a,b,c,d,f,g), simplify=False)
sol[g] = 1
```

```python
# Get rotation angle
phi_inv = (2*sol[b]/(sol[a]-sol[c])).cancel().simplify()
phi_solved = sp.Rational(1,2)*sp.atan2(sp.numer(phi_inv), sp.denom(phi_inv)); display(phi_solved)
```

```python
# Make sure we really have an ellipse
#display(sol[b]**2 - sol[a]*sol[c])

# X and Y offsets
x0 = (sol[c]*sol[d] - sol[b]*sol[f])/(sol[b]**2 - sol[a]*sol[c])
y0 = (sol[a]*sol[f] - sol[b]*sol[d])/(sol[b]**2 - sol[a]*sol[c])

x0 = x0.cancel().simplify()
y0 = y0.cancel().simplify()

x0, y0
```

Ok, so now we have the offsets and they match to the symbolic values. Now, I can get a simplified form for this stuff by subtracting out the offset terms and resolve for the coefficients. Perhaps rotating the coordinate axis will help too.

```python
Fxy_new = (a*x**2 + 2*b*x*y + c*y**2 + 2*d*x + 2*f*y + g).subs({x: I-x0, y: Q-y0}).expand(trig=True);
Fxy_new = Fxy_new.replace(sp.cos(p)**2, (1 + sp.cos(2*p))/2).replace(sp.sin(p)**2, (1 - sp.cos(2*p))/2).expand();
display(Fxy_new)
# Obtain coefficients and group terms
A = Fxy_new.coeff(sp.sin(2*w*tau))
B = Fxy_new.coeff(sp.cos(2*w*tau))
C = Fxy_new.coeff(sp.cos(w*tau)*sp.sin(w*tau))

D = Fxy_new.coeff(sp.sin(w*tau)).coeff(sp.cos(w*tau), n=0)
F = Fxy_new.coeff(sp.cos(w*tau)).coeff(sp.sin(w*tau), n=0)
G = Fxy_new.coeff(sp.cos(w*tau), n=0).coeff(sp.sin(w*tau), n=0).coeff(sp.cos(2*w*tau), n=0).coeff(sp.sin(2*w*tau), n=0)

sol2 = sp.solve([x for x in [A,B,C,D,F,G]], (a,b,c,d,f,g), simplify=True)
sol2[g] = 1
```

```python
L1 = sp.sqrt(2*(sol2[a]*sol2[f]**2 + sol2[c]*sol2[d]**2 + sol2[g]*sol2[b]**2 - 2*sol2[b]*sol2[d]*sol2[f] - sol2[a]*sol2[c]*sol2[g])/ \
             ((sol2[b]**2 - sol2[a]*sol2[c])*(sp.sqrt((sol2[a]-sol2[c])**2 + 4*sol2[b]**2) - (sol2[a] + sol2[c]))))
L2 = sp.sqrt(2*(sol2[a]*sol2[f]**2 + sol2[c]*sol2[d]**2 + sol2[g]*sol2[b]**2 - 2*sol2[b]*sol2[d]*sol2[f] - sol2[a]*sol2[c]*sol2[g])/ \
             ((sol2[b]**2 - sol2[a]*sol2[c])*(-sp.sqrt((sol2[a]-sol2[c])**2 + 4*sol2[b]**2) - (sol2[a] + sol2[c]))))
```

```python
tmp1 = (L1**2).expand(trig=True).subs({g: 1}).simplify()
tmp2 = (L2**2).expand(trig=True).subs({g: 1}).simplify()
tmp1 = tmp1.replace(sp.cos(p)**2, (1 + sp.cos(2*p))/2).replace(sp.sin(p)**2, (1 - sp.cos(2*p))/2).expand().factor()
tmp2 = tmp2.replace(sp.cos(p)**2, (1 + sp.cos(2*p))/2).replace(sp.sin(p)**2, (1 - sp.cos(2*p))/2).expand().factor()
phi_inv = (2*sol[b]/(sol[a]-sol[c])).cancel().simplify()
phi_solved = sp.Rational(1,2)*sp.atan2(sp.numer(phi_inv), sp.denom(phi_inv)); 
display(tmp1)
display(tmp2)
display(phi_solved)
```

Now that I have these analytic expressions, solve for $\delta$ and do the mapping to the circle. $E_p$ is the product of field magnitudes. $\Theta$ is $\tan(2\Phi)$.

```python
Ep,La,Lb,Phi,Theta = sp.symbols('E_p,L_a,L_b,Phi,Theta', positive=True, real=True)
tmp1 = tmp1.collect(E1**2*E2**2).subs(E1**2*E2**2, Ep**2)
tmp2 = tmp2.collect(E1**2*E2**2).subs(E1**2*E2**2, Ep**2)

res = sp.solve([phi_solved - Phi], sp.cos(delta), dict=True)
display(res[1][sp.cos(delta)].replace(sp.tan(2*Phi), Theta))
subs = {delta: sp.acos(res[1][sp.cos(delta)]).replace(sp.tan(2*Phi), Theta)}
display((tmp1.subs(subs)-La**2).expand(trig=True)) 
display((tmp2.subs(subs)-Lb**2).expand(trig=True)) 
#display(phi_solved - Phi)
```

The denominator is actually factorizable, but sympy doesn't do it automatically yet. Let's try factoring the denominator and rebuilding the expression.

```python
def factor_sqrt_args(expr):
    p = sp.Wild('p')
    subs = {}
    for i in expr.find(sp.sqrt(p)):
        subs[i] = sp.sqrt(i.args[0].factor())
    return expr.xreplace(subs)
```

I'm going to assume that $A_1 > A_2$. This gets rid of an absolute value in the denominator. I think this should be general, because $L_a$ and $L_b$ represent two roots. Choosing $A_1$ larger than $A_2$ just flips which one is which. I'm also replacing $\sqrt{\Theta^2 + 1}$ with $K$ for convenience.

```python
K = sp.symbols('K', dummy=True)
tmp3 = factor_sqrt_args((tmp1.subs(subs)-La**2).expand(trig=True)).refine(sp.Q.positive(A1 - A2)).expand().subs({sp.sqrt(Theta**2 + 1): K}).simplify()
tmp4 = factor_sqrt_args((tmp2.subs(subs)-Lb**2).expand(trig=True)).refine(sp.Q.positive(A1 - A2)).expand().subs({sp.sqrt(Theta**2 + 1): K}).simplify()
display(tmp3)
display(tmp4)
```

```python
a_subs = sp.solve([tmp3, tmp4], (A1**2, A2**2), quick=True, warning=True, dict=True)[1]
display(a_subs)
```

If I assume that $K^{2} L_{a}^{4} + 2 K^{2} L_{a}^{2} L_{b}^{2} + K^{2} L_{b}^{4} - L_{a}^{4} \Phi^{2} - L_{a}^{4} + 2 L_{a}^{2} L_{b}^{2} \Phi^{2} + 2 L_{a}^{2} L_{b}^{2} - L_{b}^{4} \Phi^{2} - L_{b}^{4}$ is nonnegative, I can combine the exponents. Then, if it is positive I can also take it out of the exponent.

```python
a_subs[A1] = sp.sqrt(a_subs[A1**2])
a_subs[A2] = sp.sqrt(a_subs[A2**2])
#a_subs[Theta] = sp.tan(2*Phi)
#del a_subs[Theta]
tmp5 = sp.solve(sp.Equality(phi_solved, Phi), sp.cos(delta))[1]
display(tmp5)
# Get the denominator in the square root.
tmp6 = sp.denom(a_subs[A1**2]).coeff(Ep**2)
#display(sp.powdenest(sp.powsimp(tmp5.subs(a_subs).expand().simplify(), force=True), force=True))
ans = sp.refine(sp.powdenest(sp.powsimp(tmp5.subs(a_subs).expand().simplify(), force=True), force=True), sp.Q.positive(tmp6))
display(ans.subs({K: sp.sqrt(Theta**2 + 1)}))
```

```python
final_ans = sp.acos(ans.subs({K: sp.sqrt(Theta**2 + 1)}).expand().simplify().factor().collect(Theta).replace(La**4 + 2*La**2*Lb**2+Lb**4, (La**2 + Lb**2)**2)).subs({Theta: sp.tan(2*Phi)}) 
display(final_ans)
```

After all these assumptions and simplifications, I get this answer for the calculated phase between $I$ and $Q$. Magically, the answer does not depend on $E_p$. 

$$
\delta = \operatorname{acos}{\left (- \frac{\left(L_{a} - L_{b}\right) \left(L_{a} + L_{b}\right) \tan{\left (2 \Phi \right )}}{\sqrt{4 L_{a}^{2} L_{b}^{2} + \left(L_{a}^{2} + L_{b}^{2}\right)^{2} \tan^{2}{\left (2 \Phi \right )}}} \right )}
$$

## Using the quadratic forms

The fitting algorithm in the Fitzgibbon minimizes the quadratic form $F(x,y) = a x^2 + b xy + c y^2 + dx + ey + f$. I want to know the geometric parameters like length and center convert to these algebraic parameters. To do this, you have to get $F(x,y)$ into standard form, which basically involves completing the square, assuming $b$ is zero. If that is not the case, the ellipse is rotated and you must rotate it back. The angle of rotation is $\arctan\left(\frac{b}{a - c}\right)$.

The standard form of the ellipse looks like $F(x,y) = \frac{(x-h)^2}{a^2} + \frac{(y-k)^2}{b^2} - 1 = 0$. It is centered at ($h$, $k$) and has a major and minor axes $2a$ and $2b$.

Furthermore, it uses the fact that rescaling the parameters by $\alpha$ doesn't change the fitting, so I have to unscale them. How do I convert these parameters to geometric parameters of the ellipse?

## Problem 4

Graphical solution of the set

$$
\begin{eqnarray}
\gamma a & = & \begin{cases}
ha\tan(ha) & \text{ for symmetric mode }\\
-ha\cot(ha) & \text{ for antisymmetric mode } \\ 
\end{cases} \\ 
(ha)^2 + (\gamma a)^2 & = & (n_2^2 - n_1^2)(k_0a)^2 \equiv V^2 \\
\end{eqnarray}
$$

for the case $n_1 = 1.480$, $n_2 = 1.495$, $\lambda_0 = 1.03 \text{$\mu$m}$, and $2a = 18\text{$\mu$m}$.

```python
%pylab inline
matplotlib.rcParams['figure.figsize'] = (12, 10)
```

```python
ha = arange(0, 12, pi/1000)
n1, n2 = 1.480, 1.495
c = 2.99e8 * 1e6 # in micron
a = 18/2
ka = 2*pi*a/1.03
V = sqrt((n2**2 - n1**2)*ka**2)
```

```python
plot(ha, ha*tan(ha), '--', label='even')
plot(ha, -ha*1/tan(ha),'-.', label='odd')

gca().add_patch(matplotlib.patches.Circle((0, 0), radius=V, fill=False))
xlim([0, 12])
ylim([0, 12])
gca().set_aspect('equal')
legend()
print 'Should get ', int(1 + floor(2*V/pi)), ' solutions'
```

```python
# There are 8 solutions
sol = array([
    [1.44638, 11.5079],
    [2.88968, 11.2331],
    [4.32981, 10.7595],
    [5.7631, 10.065],
    [7.1856, 9.1038],
    [8.5907, 7.79356],
    [9.96266, 5.9385],
    [11.2428, 2.84969]])
```

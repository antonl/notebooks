```python
%pylab inline
from nudie import SpeFile, mount_point, data_folder, simple_wavelength_axis
```

```python
data_file = SpeFile(mount_point / '2D/Anton' / 'wedge_calib-spec-0001-rephasing.spe')
data_file.header.datatype = 3 # fix data type
data = data_file.data[0:, :, 10:].squeeze()
wl = simple_wavelength_axis(groove_density=600, center_wavelength=800)
```

```python
rcParams['figure.figsize'] = (16, 12)
```

```python
plot(wl, data[:, 100])
xlim(780, 820)
grid()
```

```python
from matplotlib import animation
fig = figure()
line, = plot(wl, data[:, 100], linewidth=3)
xlim(780, 820)
grid()

def animate(i):
    line.set_ydata(data[:, i])
    return line,

anim = animation.FuncAnimation(fig, animate, arange(2000, 4000))
anim.save('basic_animation.mp4', fps=50, bitrate=20000)
```

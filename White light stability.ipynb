```python
from __future__ import division
import sys
sys.path.append('/Users/aloukian/Documents/pyWinSpec/')
from scipy.stats import variation
from winspec import *
import itertools as it
%pylab inline
matplotlib.rcParams['figure.figsize'] = (16, 16)
```

```python
wl_data = squeeze(SpeFile('/Users/aloukian/Documents/ogilvielab/Data/14-02-25/stability02.SPE').data)
```

```python
bad_pixels_big = [150, 199, 381, 584, 684, 784, 868, 927, 976]
bad_pixels_small = [246, 295, 549, 635, 847, 1003]

mask = zeros_like(wl_data, dtype=bool8)
mask[:, list(it.chain([i for px in bad_pixels_big for i in xrange(px-4, px+8)], 
                              [i for px in bad_pixels_small for i in xrange(px-4, px+5)]))] = True
wl_masked = ma.masked_array(wl_data, mask=mask)
```

```python
fig, (ax1, ax2, ax3) = subplots(3, 1, sharex=True)
ax1.plot(variation(wl_data), color='red')
ax1.plot(ma.filled(ma.std(wl_masked, axis=0)/ma.mean(wl_masked, axis=0), nan))
ax2.plot(mean(wl_data, axis=0), color='red')
ax2.plot(ma.mean(wl_masked, axis=0))
ax1.grid()
ax2.grid()
ax1.set_title('variation')

ax3.plot(variation(diff(wl_data, axis=0)))
```

```python
def scan_fdm(signal, dt, intervals=None, a_thresh=1e-2, rel_err=1e-4):
    N = signal.shape[0]
    f_res = 1/(N*dt)
    if intervals is None:
        intervals = int(N/10)
    solutions = []
    
    for fmin,fmax in [(float(i-1)/intervals/(4*dt), float(i)/intervals/(4*dt)) for i in xrange(1, intervals+1)]:
        #print fmin, fmax
        z, A = fdm.solve_fdm(fdm.make_basis(fmin, fmax, T=N*dt, dt=dt, tweak=0.6), signal)
        z, A, errs = fdm.solve_fdm(exp(-1j*2*pi*real(z)), signal,
        amplitude_threshold=a_thresh, relerr_threshold=rel_err, gen_u2=True)
        solutions.append((z/dt, A, errs))
    
    zs = np.zeros((0,), dtype=complex128)
    errs = np.zeros((0,), dtype=float64)
    amps = np.zeros((0,), dtype=float64)
    for (z, A, err) in solutions:
        zs = append(zs, z)
        errs = append(errs, err)
        amps = append(amps, A)
    return (zs, errs, amps)

def kmeans(n_clusters=None):
    model = sklearn.cluster.KMeans(n_clusters=4)
    color_iter = it.cycle(['r', 'g', 'b', 'c', 'm'])
    X = vstack((real(zs), imag(zs))).T
    model.fit(X)
    A = zeros((model.n_clusters,), dtype='complex128')
    for i,c in it.izip(unique(model.labels_), color_iter):
        idx = argwhere(model.labels_ == i)
        A[i] = Amps[idx].mean() 
        scatter(X[idx, 0], X[idx, 1], color=c)
```
